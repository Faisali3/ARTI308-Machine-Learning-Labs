# Lab 4: Data Quality Assessment & Preprocessing (With Visualizations)

## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.decomposition import PCA
```

## 2. Load Dataset

```python
df = pd.read_csv("ev_vs_petrol_dataset_v3.csv")
df.head()
```

## 3. Data Exploration

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

## 4. Distribution of Amount

```python
plt.figure()
plt.hist(df['Amount'], bins=20)
plt.title("Distribution of Amount")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.show()
```

## 5. Boxplot for Outlier Detection

```python
plt.figure()
plt.boxplot(df['Amount'])
plt.title("Boxplot of Amount")
plt.show()
```

## 6. Outlier Detection Using IQR

```python
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

lower_bound, upper_bound
```

## 7. Normalization

```python
df_minmax = df[['Amount', 'Boxes Shipped']].copy()

minmax = MinMaxScaler()
df_minmax[['Amount', 'Boxes Shipped']] = minmax.fit_transform(df_minmax)

df_minmax.head()
```

```python
df_zscore = df[['Amount', 'Boxes Shipped']].copy()

standard = StandardScaler()
df_zscore[['Amount', 'Boxes Shipped']] = standard.fit_transform(df_zscore)

df_zscore.head()
```

## 8. PCA

```python
X = df_zscore[['Amount', 'Boxes Shipped']]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X)

pca.explained_variance_ratio_
```

## 9. PCA Scatter Plot

```python
plt.figure()
plt.scatter(principal_components[:,0], principal_components[:,1])
plt.title("PCA Projection")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
